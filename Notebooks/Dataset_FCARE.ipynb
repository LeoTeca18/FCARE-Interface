Carregamento do Dataset

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("mlg-ulb/creditcardfraud")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta
import seaborn as sns
import matplotlib.pyplot as plt


# Carregar os dados
df = pd.read_csv(path + "/creditcard.csv")

# Ver as primeiras linhas
df
```

```python

# Lista completa das 18 províncias de Angola
provincias_angola = [
    "Bengo", "Benguela", "Bié", "Cabinda", "Cuando Cubango", "Cuanza Norte", "Cuanza Sul",
    "Cunene", "Huambo", "Huíla", "Luanda", "Lunda Norte", "Lunda Sul", "Malanje",
    "Moxico", "Namibe", "Uíge", "Zaire", "Moxico Leste", "Cuando", "Icolo e Bengo"
]
df["localizacao"] = random.choices(provincias_angola, k=len(df))
df
```

```python
# Categorias de compra mais detalhadas
categorias_compra = [
    "Supermercado", "Eletrónicos", "Vestuário", "Restauração", "Transporte",
    "Saúde", "Educação", "Hotelaria", "Lazer", "Telecomunicações", "Combustível",
    "Automóveis", "Mobiliário", "Serviços Financeiros"
]
df["categoria_compra"] = random.choices(categorias_compra, k=len(df))
df
```

```python
df["tempo_HHMMSS"] = pd.to_timedelta(df["Time"], unit="s")
df
```

```python
#Time e Renomear 'Amount' para 'valor'
df = df.rename(columns={"Amount": "valor_gasto", "Class": "classe", "Time": "tempo"})
df
```

```python
# Converter valores para Kwanza (KZ) e arredondar
df["valor_gasto"] = df["valor_gasto"] * 825  # Supondo que 1 unidade do dataset original ≈ 825 KZ
df["valor_gasto"] = df["valor_gasto"].round(2)  # Arredondar para 2 casas decimais

# Aplicar formatação de valor para separador de milhar (Exemplo: "123 456.5 KZ")
df["valor_gasto"] = df["valor_gasto"].apply(lambda x: f"{x:,.1f}".replace(",", " ").replace(".", ",")) + " KZ"
df
```

```python
import random

# Lista de nomes fixos para os usuários
nomes = [
    "Carlos", "Maria", "José", "Ana", "Pedro", "Fátima", "João", "Teresa",
    "Francisco", "Marta", "Alexandre", "Isabel", "Manuel", "Joana", "Ricardo",
    "Beatriz", "Sérgio", "Helena", "Rui", "Patrícia", "Gonçalo", "Andreia",
    "Fernando", "Tatiana", "Bruno", "Cátia", "Paulo", "Larissa", "Vitor",
    "Tatiane", "Leonel", "Luana", "Emanuel", "Daniela", "Hugo", "Carla",
    "Marcos", "Vanessa", "Nelson", "Adriana", "Duarte", "Raquel", "Orlando",
    "Camila", "Amílcar", "Miguel", "Sofia", "Jorge", "Cristina"
]

tipos_transacao = ["Compra online", "Levantamento", "Transferência", "Pagamento de serviço"]

# Adicionar as coluna ao dataframe
df["tipo_transacao"] = [random.choice(tipos_transacao) for _ in range(len(df))]

# Criar uma lista de IDs únicos correspondendo aos nomes
ids_unicos = list(range(1, len(nomes) + 1))  # IDs de 1 até o número de nomes

# Criar um dicionário fixo associando cada ID a um nome específico
id_para_nome = dict(zip(ids_unicos, nomes))

# Atribuir IDs ao dataset de forma aleatória, mas garantindo nomes fixos
df["id"] = [random.choice(ids_unicos) for _ in range(len(df))]
df["nome"] = df["id"].map(id_para_nome)  # Garantir que cada ID tenha sempre o mesmo nome

# Exibir as primeiras linhas do dataset atualizado
df
```

```python
# Bancos emissores presentes em Angola
bancos_angola = [
    "Banco Angolano de Investimentos (BAI)", "Banco de Fomento Angola (BFA)",
    "Banco Económico", "Banco Millennium Atlântico", "Banco BIC", "Banco Keve",
    "Banco Sol", "Standard Bank Angola", "Banco BNI", "Banco Yetu"
]
df["banco_emissor"] = [random.choice(bancos_angola) for _ in range(len(df))]
df
```

```python
# Reorganizando a ordem das colunas
#df = df[["id", "nome", "valor_gasto", "categoria_compra", "tempo_HHMMSS", "tipo_transacao", "localizacao", "banco_emissor", "classe"]]

# Função para limpar e converter 'valor_gasto' para float
def clean_amount(value):
    if isinstance(value, str):  # Garantir que é string antes do processamento
        value = value.replace("KZ", "").replace(" ", "").replace(",", ".")  # Remover 'KZ', espaços e ajustar decimal
    try:
        return float(value)
    except ValueError:
        return None  # Se houver erro, retorna None para tratamento posterior

# Aplicar limpeza na coluna 'valor_gasto'
df.loc[:, "valor_gasto"] = df["valor_gasto"].apply(clean_amount)

# Ajustar a extração da hora de 'tempo_HHMMSS'
df.loc[:, "tempo_HHMMSS"] = df["tempo_HHMMSS"].astype(str).str.extract(r'(\d{2}):(\d{2}):(\d{2})')[0]

# Converter para inteiro, garantindo que os valores não sejam nulos
df["tempo_HHMMSS"] = pd.to_numeric(df["tempo_HHMMSS"], errors='coerce').fillna(0).astype(int)
df.rename(columns={"tempo_HHMMSS":"hora"}, inplace=True)
df
```
